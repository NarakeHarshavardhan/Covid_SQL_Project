```python
import pandas as pd

# Set the maximum number of rows to display
pd.set_option('display.max_rows', 1000)
```

```python
# Load first Excel file into a DataFrame
CovidDeaths = pd.read_excel('CovidDeaths.xlsx')

# Load second Excel file into a DataFrame
CovidVaccinations = pd.read_excel('CovidVaccinations.xlsx')
```

```python
# Create a SQLAlchemy engine
engine = create_engine('sqlite:///PortfolioProject.db')

# Connect to the database
conn = engine.connect()
```

```python
# Insert first DataFrame into the database as a table
CovidDeaths.to_sql('CovidDeaths', conn, index=False, if_exists='replace')

# Insert second DataFrame into the database as a table
CovidVaccinations.to_sql('CovidVaccinations', conn, index=False, if_exists='replace')
```

```python
q12 = pd.read_sql("""
    SELECT dea.continent, dea.location, dea.population, vac.new_vaccinations,
    SUM(vac.new_vaccinations) OVER (PARTITION BY dea.location ORDER BY dea.date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS rolling_sum
    FROM CovidDeaths dea
    JOIN CovidVaccinations vac
    ON dea.location = vac.location
    AND dea.date = vac.date
    WHERE dea.continent IS NOT NULL
    ORDER BY dea.location, dea.date
    """, conn)
q12.head(500)
```
